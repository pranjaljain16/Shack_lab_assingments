```python
import numpy as np 
import pandas as pd
import os
import warnings
warnings.filterwarnings('ignore')
```

## READING FLIPKART DATASET

```python
df_f = pd.read_csv('flipkart_com-ecommerce_sample.csv')
```

```python
df_f.head()
```

## READING AMAZON DATASET

```python
df_a = pd.read_csv('amz_com-ecommerce_sample.csv',encoding= 'unicode_escape')
```

```python
df_a.head()
```

```python
df_a.shape
```

```python
df_f.shape
```

## MERGING THE TABLES ON UNIQ_ID COLUMN

```python
df = df_f.merge(df_a, on='uniq_id', how='left')
```

```python
df.head()
```

```python
df.shape
```

## CREATING A DIFFRENT DATASET WITH NAME , RETAIL PRICE AND DISCOUNTED PRICE

```python
df1 = df[['product_name_x', 'retail_price_x', 'discounted_price_x' , 'product_name_y' , 'retail_price_y', 'discounted_price_y']]
```

```python
df1.head()
```

### RENAMING THE COLUMNS

```python
final_df = df1.rename(columns = {'product_name_x':'Product name in Flipkart', 'retail_price_x':'Retail Price in Flipkart',
           'discounted_price_x':'Discounted Price in Flipkart','product_name_y':'Product name in Amazon'
           ,'retail_price_y':'Retail Price in Amazon','discounted_price_y':'Discounted Price in Amazon'})
```
